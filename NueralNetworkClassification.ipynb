```python
#Andres Cervantes
#Sources used:
#https://machinelearningmastery.com/display-deep-learning-model-training-history-in-keras/
#https://www.pluralsight.com/guides/classification-keras
#Jupyter Lecture notebook
#https://www.tensorflow.org/api_docs/python/tf/keras/Sequential

# Import required libraries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import sklearn

# Import necessary modules
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math import sqrt

# Keras specific
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical 
from keras import layers
from tensorflow import keras
```

```python
# Load the dataset as numpy 

ionoTrain = np.genfromtxt("data/ionosphere/preprocessed_ionosphere_train.csv", delimiter=',')
ionoTest = np.genfromtxt("data/ionosphere/preprocessed_ionosphere_test.csv", delimiter=',')

# Separate features from target (Complete the code)
X_train = ionoTrain[:,0:34]
y_train = ionoTrain[:,34]

X_test = ionoTest[:,0:34]
y_test = ionoTest[:,34]

#print(X_train.shape)
#print(y_train.shape)
#print(X_test.shape)
#print(y_train.shape)

y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

model = Sequential()
model.add(Dense(34, activation='relu', input_dim=34))
model.add(Dense(34, activation='relu'))
model.add(Dense(17, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(2, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', 
              loss='binary_crossentropy', 
              metrics=['accuracy'])

history = model.fit(X_train, y_train, validation_data=(X_test, y_test), batch_size=128, epochs=100, verbose=0)

print("\n\nTRAINING ACCCURACY ------------------------")
print(history.history['accuracy'])

print("\n\nVALIDATION ACCCURACY ------------------------")
print(history.history['val_accuracy'])


print("\n\nCompared to my training and validation accuracy from HW 4, this model is\n"+
"significantly more accurate than using logistic regression. In the Logistic Regression model of HW4\n"+
"the highest value in the training and validation accuracy was about .90. Here, in the NN model, the\n"+
"highest values of accuracy of both the training and validation data range from .97 to .99 after several epochs.")

# summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

#END QUESTION 1--------------------------------------------------------------------------------------------




```

```python
#BEGIN PROBLEM 2 -------------------------------------------------------------------



(X_train, y_train), (X_test, y_test) = keras.datasets.cifar10.load_data()
assert X_train.shape == (50000, 32, 32, 3)
assert X_test.shape == (10000, 32, 32, 3)
assert y_train.shape == (50000, 1)
assert y_test.shape == (10000, 1)



input_shape = (32,32,3)

input_shape_flat = (input_shape[0] * input_shape[1] * input_shape[2],)
print(input_shape_flat)

X_train = np.reshape(X_train, (-1,input_shape_flat[0])) # -1 means let reshape determines the value of this input
X_test = np.reshape(X_test, (-1,input_shape_flat[0]))   # -1 means let reshape determines the value of this input
print("x_train shape:", X_train.shape)
print("x_test shape:", X_test.shape)

# Scale images to the [0, 1] range. 256 gray levels per pixel
X_train = X_train.astype("float32") / 255
X_test = X_test.astype("float32") / 255

# Model / data parameters
num_classes = 10
# convert class vectors to binary class matrices
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)

L1 = layers.Dense(1536, activation='relu') # 24 neurons -> 24 outputs; ReLU, i.e. max{x,0}, activation
L2 = layers.Dense(768, activation='relu') # can specify no activation
L3 = layers.Dense(384, activation='relu') # 24 neurons -> 24 outputs; ReLU, i.e. max{x,0}, activation
L4 = layers.Dense(10, activation='softmax') # can specify no activation

inLayer = layers.Input(shape=input_shape_flat)

nnOutput = L4(L3(L2(L1(inLayer))))

nn = keras.models.Model(inputs=inLayer,outputs=nnOutput)

nn.compile(loss="categorical_crossentropy", optimizer="adam", metrics=['accuracy'])
log = nn.fit( X_train, y_train, validation_data=(X_test, y_test), batch_size=128, epochs=100)

print("\n\nTRAINING ACCCURACY ------------------------")
print(log.history['accuracy'])

print("\n\nVALIDATION ACCCURACY ------------------------")
print(log.history['val_accuracy'])

print("\n\nBased on the data, the validation accuracy is not very good compared to the test data since it averages at about 0.5.")

# summarize history for accuracy
plt.plot(log.history['accuracy'])
plt.plot(log.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

# summarize history for loss
plt.plot(log.history['loss'])
plt.plot(log.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
